```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
yelp = pd.read_csv('yelp_review_cleaned.csv')
```

```python
data = yelp.text_lower_no_digits_non_punct_split_non_stopwords_stemmed[:1000]
tfidf_vec = TfidfVectorizer()
X1 = tfidf_vec.fit_transform(data)
unigrams = tfidf_vec.get_feature_names_out()

## All words in the vocabulary.
# print("All words in the vocabulary:",unigrams)
```

```python
dtm1 = pd.DataFrame(X1.toarray(), columns=unigrams)
dtm1.head()
```

### (1) Printing the unigrams with their tf-idf values in descending order

```python
tf_idf_values = dict(dtm1.sum(axis = 0))
sorted(tf_idf_values.items(), key=lambda x: x[1], reverse=True)
```

```python
tfidf_vec = TfidfVectorizer(ngram_range=(1,3))
X123 = tfidf_vec.fit_transform(data)
```

```python
dtm123 = pd.DataFrame(X123.toarray(), columns=tfidf_vec.get_feature_names_out())
dtm123
```

### (2) Printing the top terms with their TF-IDF values in descending order

```python
tf_idf_values = dict(dtm123.sum(axis = 0))
sorted(tf_idf_values.items(), key=lambda x: x[1], reverse=True)[:3]
```

### (3) According to the results in (2), do you think these terms and their higher TF-IDF values make sense in the Yelp review context?

The top three terms, `food`, `good`, `great` definitely make sense in the context of the Yelp dataset, as this data set handles reviews of food places.
