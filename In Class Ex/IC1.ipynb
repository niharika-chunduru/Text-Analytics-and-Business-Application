```python
fl_business = pd.read_csv(r"/content/drive/MyDrive/Sem-IV/INFOSYS723 - Text Analytics and Business Application/fl_business.csv")

fl_business.head()
```

1. In `attributes_BusinessAcceptsCreditCards` column, fill all missing values as 'None'.

```python
# The fillna() function replaces any nan values in a Pandas Series with the value specified.
# The inplace=True flag ensures that the original Series is modified instead of creating a new Series.
fl_business.attributes_BusinessAcceptsCreditCards.fillna('None', inplace=True)
```

```python
fl_business.attributes_BusinessAcceptsCreditCards.value_counts()
```

2. In `attributes_BusinessAcceptsCreditCards` column, please convert all 'TRUE' and True values as 1, and all other values as 0. Store all converted values into a new column called `AcceptCreditCards`.

```python
# The replace() function accepts a dictionary of values where
# The keys are the elements of the Series that need to be replace, and
# the values are the elements that they are replaced with.
fl_business['AcceptCreditCards'] = fl_business.attributes_BusinessAcceptsCreditCards.replace({True:1, 'TRUE':1, 'None':0, False:0, 'FALSE':0})
```

```python
fl_business.AcceptCreditCards.value_counts()
```

3. Please drop rows that have missing values in column `attributes_RestaurantsPriceRange2`.

```python
# The dropna() function removes missing values.
# The subset paramenter allows to choose what column to look for nan values.
# The inplace=True flag ensures that the original DataFrame is modified instead of creating a new DataFrame.
fl_business.dropna(subset=['attributes_RestaurantsPriceRange2'],inplace=True)
```

```python
fl_business.attributes_RestaurantsPriceRange2.value_counts()
```

4. Column `attributes_RestaurantsCounterService` has a lot of missing data. Please drop this column.

```python
# The drop() function removes rows or column by index.
# The label value contains the name of the column that needs to be dropped.
# The axis=1 flag indicates that a column needs to be removed.
# The inplace=True flag ensures that the original DataFrame is modified instead of creating a new DataFrame.

fl_business.drop('attributes_RestaurantsCounterService',axis=1,inplace=True)
```
